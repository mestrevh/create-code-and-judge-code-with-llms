### Resultados GPT
---

```python
database = pd.read_csv('result_send_code_gpt.csv')
```

```python
empty = []
presentation_error = []
compilation_error = []
runtime_error = []

for i in range(len(database)):
    if database['evaluation'][i] == 'EMPTY_ANSWER':
        empty.append(f"id: {database['id'][i]} - {database['name'][i]}")
    if database['evaluation'][i] == 'PRESENTATION_ERROR':
        presentation_error.append(f"id: {database['id'][i]} - {database['name'][i]}")
    if database['evaluation'][i] == 'RUNTIME_ERROR':
        runtime_error.append(f"id: {database['id'][i]} - {database['name'][i]}")
    if database['evaluation'][i] == 'COMPILATION_ERROR':
        compilation_error.append(f"id: {database['id'][i]} - {database['name'][i]}")
```

```python
def print_questions_error(data: list, title: str):
    print(f"Questões com tipo de erro: {title}\n")
    
    for value in data:
        print(value)
```

```python
print_questions_error(empty, "Empty Answer")
```

```python
print_questions_error(presentation_error, "Presentation Error")
```

```python
print_questions_error(compilation_error, "Compilation Error")
```

```python
print_questions_error(runtime_error, "Runtime Error")
```

```python
count_evaluation = {}
count_level = {}

for i in range(0, len(database['evaluation'])):
    try:
        count_evaluation[database['evaluation'][i]] += 1
    except:
        count_evaluation[database['evaluation'][i]] = 0
        count_evaluation[database['evaluation'][i]] += 1
        
    try:
        count_level[f"Level {int(database['level'][i])}"] += 1
    except:
        count_level[f"Level {int(database['level'][i])}"] = 0
        count_level[f"Level {int(database['level'][i])}"] += 1

print(count_evaluation)
print(count_level)
```

```python
def plot_graphic(columns: list, values: list, title: str = "", label_x="", label_y="", color="blue"):
    plt.figure(figsize=(8, 5))  # tamanho do gráfico
    bars = plt.barh(columns, values, color=color)

    # Adiciona valores no final de cada barra
    for bar in bars:
        width = bar.get_width()
        plt.text(width + (max(values) * 0.01),  # desloca um pouquinho para frente
                 bar.get_y() + bar.get_height()/2,
                 f"{width}",
                 va="center")

    plt.title(title)
    plt.xlabel(label_x)
    plt.ylabel(label_y)
    plt.tight_layout()
    plt.show()
```

```python
plot_graphic(count_level.keys(), count_level.values(), "Quantidade de questões por level", label_y="", label_x="Quantidade")
```

```python
plot_graphic(count_evaluation.keys(), count_evaluation.values(), "Respostas ao enviar código do Gemini", label_x="")
```
